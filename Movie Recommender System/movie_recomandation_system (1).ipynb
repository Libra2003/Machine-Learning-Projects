```python
# this lines merges 2 data sets together
movies = movies.merge(credits,on='title')
```

```python
# geners
# id
# keywords
# title
# overview
# cast
# crew
# Reassigning number of necessary columns to the movie dataset
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
#This line of code is used to detect missing values form the
movies.isnull().sum()
```

```python
#This function is used to drop any missing values from the dataframe
movies.dropna(inplace=True)
```

```python
# This function checks if there are any duplicate values in the data franme
movies.duplicated().sum()
```

```python
#This function displays
movies.iloc[0].genres
```

```python
#Fuction to append names of the geners to a single list
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert2(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])

  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['crew'][0]
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#New dataFrame for holding these values
new_movies_df = movies[['id', 'title','tags']]
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#This function is used to for stemming
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_movies_df['tags'][0]
```

```python
new_movies_df['tags'] = new_movies_df['tags'].apply(stem)
```

```python
new_movies_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies_df['tags']).toarray()
```
